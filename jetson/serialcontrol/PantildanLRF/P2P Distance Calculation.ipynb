```python
import math
```

```python
# Given data
range1 = 7.1  # meters for both points
range2 = 6.7
```

```python
# Yaw and Pitch angles in degrees
yaw1_deg = -18.5
pitch1_deg = 0
yaw2_deg = -2
pitch2_deg = 5.9
```

```python
# Convert degrees to radians
yaw1_rad = math.radians(yaw1_deg)
pitch1_rad = math.radians(pitch1_deg)
yaw2_rad = math.radians(yaw2_deg)
pitch2_rad = math.radians(pitch2_deg)

# Cartesian coordinates for Point 1
x1 = range1 * math.cos(pitch1_rad) * math.cos(yaw1_rad)
y1 = range1 * math.cos(pitch1_rad) * math.sin(yaw1_rad)
z1 = range1 * math.sin(pitch1_rad)

# Cartesian coordinates for Point 2
x2 = range2 * math.cos(pitch2_rad) * math.cos(yaw2_rad)
y2 = range2 * math.cos(pitch2_rad) * math.sin(yaw2_rad)
z2 = range2 * math.sin(pitch2_rad)
```

```python
# Euclidean distance between Point 1 and Point 2
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)

distance
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the two points
ax.scatter(x1, y1, z1, color='r', label='Point 1')
ax.scatter(x2, y2, z2, color='b', label='Point 2')

# Draw a line between Point 1 and Point 2
ax.plot([x1, x2], [y1, y2], [z1, z2], color='g', label='Distance Line')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
